1) What are the top 10 highest-rated iPhones on Flipkart in India?

```python
highest_rated = data.sort_values(by = ['Star Rating'],ascending = False)

highest_rated = highest_rated.head(10)
```

```python
print(highest_rated['Product Name'])
```

2) How many ratings do the highest-rated iPhones on Flipkart have?

```python
data.head()
```

```python
iphones = highest_rated['Product Name'].value_counts()
label = iphones.index
counts = highest_rated['Number Of Ratings']
figure = px.bar(highest_rated,x=label, y = counts)
figure.show()
```

3) Which iPhone has the highest number of reviews on Flipkart?

```python
data.head()
```

```python
highest_rated2 = data.sort_values(by = ['Number Of Ratings'],ascending = False)

highest_rated2 = highest_rated2.head(10)
```

```python
iphones = highest_rated2['Product Name'].value_counts()
label = iphones.index
counts = highest_rated2['Number Of Reviews']
figure = px.bar(highest_rated2,x=label, y = counts,title = "Number of reviews of highest rated iphones")
figure.show()
```

4) What is the relationship between the sale price of iPhones and the number of ratings on Flipkart?

```python
data.head()
```

```python
figure = px.scatter(data_frame = data,x="Number Of Ratings", y="Sale Price", size ='Discount Percentage', trendline ='ols', title=" Relationship between sales price and number of rating ")
figure.show()
```

5) What is the relationship between the discount percentage and the number of ratings of iPhones on Flipkart?


```python
figure = px.scatter(data_frame = data , x = "Number Of Ratings", y ="Discount Percentage",size ="Sale Price",trendline ='ols', title ="Relationship between disciunt percentage and Number of ratings " )
figure.show()
```

6) Can you figure out the least expensive and most expensive iPhones in the Indian market, along with all their specifications?

```python
data.head()
```

```python
most_expensive = data.loc[data['Sale Price'].idxmax()]
least_expensive = data.loc[data['Sale Price'].idxmin()]

print("Most Expensive iPhone")
print(most_expensive)

print("\nLeast Expensive iPhone")
print(least_expensive)
```
